```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set plotting style
plt.style.use('seaborn-v0_8-whitegrid')

# Configuration for plots
plt.rcParams['figure.figsize'] = (15, 8)
plt.rcParams['font.size'] = 11

print("Libraries imported successfully!")
print("Ready for Task 1 vs Task 3 comparison")
```

## 3. Side-by-Side Comparison Graphs

```python
# Create simple comparison graphs based on existing results

# Prepare data for comparison
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Task 1 vs Task 3: Wireless vs Wired Network Comparison', fontsize=16, fontweight='bold')

# Wireless data from Task 1 (based on notebook results)
wireless_configs = ['conf1', 'conf2', 'conf3', 'conf4', 'conf5', 'conf6']
wireless_hosts = [10, 30, 80, 170, 300, 400]
wireless_pdr = [97.5, 97.5, 97.5, 97.2, 60.5, 48.0]
wireless_throughput = [0.0078, 0.0234, 0.0624, 0.1323, 0.1453, 0.1537]  # Mbps
wireless_delay = [8.9, 28.8, 97.8, 233.2, 290.5, 327.2]  # ms

# Wired data from Task 3 (based on CSV)
wired_pdr = [97.5, 97.5, 97.5, 97.5, 97.5, 97.5]
wired_throughput = [0.0078, 0.0234, 0.0624, 0.1326, 0.234, 0.312]  # Mbps
wired_delay = [0.098, 0.231, 0.563, 1.160, 2.023, 2.687]  # ms
wired_loads = [16, 24, 480, 960, 1280, 1280]  # Mbps

# Plot 1: PDR Comparison
ax1.plot(wireless_hosts, wireless_pdr, 'r-o', linewidth=3, markersize=8, label='Wireless (Task 1)')
ax1.plot(range(1, 7), wired_pdr, 'b-s', linewidth=3, markersize=8, label='Wired (Task 3)')
ax1.set_xlabel('Configuration Number / Host Count', fontweight='bold')
ax1.set_ylabel('Packet Delivery Ratio (%)', fontweight='bold')
ax1.set_title('PDR: Wireless Degrades, Wired Stable', fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim([0, 105])

# Plot 2: Throughput Comparison
ax2.plot(wireless_hosts, wireless_throughput, 'r-o', linewidth=3, markersize=8, label='Wireless (Task 1)')
ax2.plot(range(1, 7), wired_throughput, 'b-s', linewidth=3, markersize=8, label='Wired (Task 3)')
ax2.set_xlabel('Configuration Number / Host Count', fontweight='bold')
ax2.set_ylabel('Throughput (Mbps)', fontweight='bold')
ax2.set_title('Throughput: Wired Scales Better', fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Delay Comparison
ax3.plot(wireless_hosts, wireless_delay, 'r-o', linewidth=3, markersize=8, label='Wireless (Task 1)')
ax3.plot(range(1, 7), wired_delay, 'b-s', linewidth=3, markersize=8, label='Wired (Task 3)')
ax3.set_xlabel('Configuration Number / Host Count', fontweight='bold')
ax3.set_ylabel('End-to-End Delay (ms)', fontweight='bold')
ax3.set_title('Delay: Wireless Shows Dramatic Increase', fontweight='bold')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Plot 4: Performance Summary Bar Chart
metrics = ['Best PDR\n(%)', 'Worst PDR\n(%)', 'Max Throughput\n(Mbps)', 'Min Delay\n(ms)', 'Max Delay\n(ms)']
wireless_values = [97.5, 48.0, 0.154, 8.9, 327.2]
wired_values = [97.5, 97.5, 0.312, 0.098, 2.687]

x = np.arange(len(metrics))
width = 0.35

bars1 = ax4.bar(x - width/2, [97.5, 48.0, 0.154, 8.9, 327.2], width, label='Wireless', alpha=0.8, color='red')
bars2 = ax4.bar(x + width/2, [97.5, 97.5, 0.312, 0.098, 2.687], width, label='Wired', alpha=0.8, color='blue')

ax4.set_xlabel('Performance Metrics', fontweight='bold')
ax4.set_ylabel('Values', fontweight='bold')
ax4.set_title('Key Performance Metrics Summary', fontweight='bold')
ax4.set_xticks(x)
ax4.set_xticklabels(metrics)
ax4.legend()
ax4.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for bar in bars1:
    height = bar.get_height()
    ax4.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

for bar in bars2:
    height = bar.get_height()
    ax4.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
