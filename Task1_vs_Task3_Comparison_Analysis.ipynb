## Objective
Simple comparison of the existing graph results from Task 1 (Wireless) and Task 3 (Wired) notebooks.

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set plotting style
plt.style.use('seaborn-v0_8-whitegrid')

# Configuration for plots
plt.rcParams['figure.figsize'] = (15, 8)
plt.rcParams['font.size'] = 11

print("Libraries imported successfully!")
print("Ready for Task 1 vs Task 3 comparison")
```

## 1. Load Data from Both Tasks
Load the key results from Task 1 (wireless) and Task 3 (wired) for comparison.

```python
# Load Task 1 (Wireless) results from existing CSV
wireless_data = pd.read_csv('task1_extracted_data_fixed.csv')

# Load Task 3 (Wired) results from existing CSV  
wired_data = pd.read_csv('task3_wired_results.csv')

print("=== DATA LOADED ===")
print(f"Wireless configurations: {len(wireless_data)}")
print(f"Wired configurations: {len(wired_data)}")

# Show sample of each
print("\n=== WIRELESS DATA SAMPLE ===")
print(wireless_data[['config', 'packet_delivery_ratio', 'throughput_bps', 'avg_end_to_end_delay']].head())

print("\n=== WIRED DATA SAMPLE ===")
print(wired_data[['configuration', 'packet_delivery_ratio', 'throughput_bps', 'avg_end_to_end_delay_ms']].head())
```

```python
# Simple comparison of key metrics
print("=== TASK 1 (WIRELESS) vs TASK 3 (WIRED) COMPARISON ===")

# Task 1 Wireless key results
print("\nTask 1 Wireless Results:")
print("- Host count range: 10 to 400 hosts")
print("- PDR degradation: 97.5% → 48% (severe saturation)")
print("- Delay increase: 8.9ms → 327ms (dramatic)")
print("- Saturation threshold: ~200 hosts")

# Task 3 Wired key results  
print("\nTask 3 Wired Results:")
print("- Consistent 97.5% PDR across all configurations")
print("- Stable delay: 0.1ms to 2.7ms (minimal variation)")
print("- No saturation observed")
print("- Predictable performance scaling")
```

## 3. Side-by-Side Comparison Graphs

```python
# Create separate side-by-side graphs for wireless and wired results

# Data from both tasks
wireless_configs = ['conf1', 'conf2', 'conf3', 'conf4', 'conf5', 'conf6']
wireless_hosts = [10, 30, 80, 170, 300, 400]
wireless_pdr = [97.5, 97.5, 97.5, 97.2, 60.5, 48.0]
wireless_throughput = [0.0078, 0.0234, 0.0624, 0.1323, 0.1453, 0.1537]  # Mbps
wireless_delay = [8.9, 28.8, 97.8, 233.2, 290.5, 327.2]  # ms

wired_configs = ['conf1', 'conf2', 'conf3', 'conf4', 'conf5', 'conf6']
wired_pdr = [97.5, 97.5, 97.5, 97.5, 97.5, 97.5]
wired_throughput = [0.0078, 0.0234, 0.0624, 0.1326, 0.234, 0.312]  # Mbps
wired_delay = [0.098, 0.231, 0.563, 1.160, 2.023, 2.687]  # ms

# Create 3 rows x 2 columns layout for side-by-side comparison
fig, axes = plt.subplots(3, 2, figsize=(18, 15))
fig.suptitle('Task 1 vs Task 3: Wireless vs Wired Network Comparison\n(Side-by-Side Analysis)', fontsize=16, fontweight='bold')

# Row 1: PDR Comparison
# Wireless PDR
axes[0,0].plot(wireless_hosts, wireless_pdr, 'r-o', linewidth=3, markersize=8)
axes[0,0].set_xlabel('Host Count', fontweight='bold')
axes[0,0].set_ylabel('Packet Delivery Ratio (%)', fontweight='bold')
axes[0,0].set_title('Task 1: Wireless PDR (Shows Saturation)', fontweight='bold', color='red')
axes[0,0].grid(True, alpha=0.3)
axes[0,0].set_ylim([0, 105])
for i, (x, y) in enumerate(zip(wireless_hosts, wireless_pdr)):
    axes[0,0].annotate(f'{y:.1f}%', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

# Wired PDR
axes[0,1].plot(range(1, 7), wired_pdr, 'b-s', linewidth=3, markersize=8)
axes[0,1].set_xlabel('Configuration Number', fontweight='bold')
axes[0,1].set_ylabel('Packet Delivery Ratio (%)', fontweight='bold')
axes[0,1].set_title('Task 3: Wired PDR (Stable Performance)', fontweight='bold', color='blue')
axes[0,1].grid(True, alpha=0.3)
axes[0,1].set_ylim([0, 105])
for i, y in enumerate(wired_pdr):
    axes[0,1].annotate(f'{y:.1f}%', (i+1, y), textcoords="offset points", xytext=(0,10), ha='center')

# Row 2: Throughput Comparison
# Wireless Throughput
axes[1,0].plot(wireless_hosts, wireless_throughput, 'r-o', linewidth=3, markersize=8)
axes[1,0].set_xlabel('Host Count', fontweight='bold')
axes[1,0].set_ylabel('Throughput (Mbps)', fontweight='bold')
axes[1,0].set_title('Task 1: Wireless Throughput (Saturates)', fontweight='bold', color='red')
axes[1,0].grid(True, alpha=0.3)
for i, (x, y) in enumerate(zip(wireless_hosts, wireless_throughput)):
    axes[1,0].annotate(f'{y:.3f}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

# Wired Throughput
axes[1,1].plot(range(1, 7), wired_throughput, 'b-s', linewidth=3, markersize=8)
axes[1,1].set_xlabel('Configuration Number', fontweight='bold')
axes[1,1].set_ylabel('Throughput (Mbps)', fontweight='bold')
axes[1,1].set_title('Task 3: Wired Throughput (Linear Scale)', fontweight='bold', color='blue')
axes[1,1].grid(True, alpha=0.3)
for i, y in enumerate(wired_throughput):
    axes[1,1].annotate(f'{y:.3f}', (i+1, y), textcoords="offset points", xytext=(0,10), ha='center')

# Row 3: Delay Comparison
# Wireless Delay
axes[2,0].plot(wireless_hosts, wireless_delay, 'r-o', linewidth=3, markersize=8)
axes[2,0].set_xlabel('Host Count', fontweight='bold')
axes[2,0].set_ylabel('End-to-End Delay (ms)', fontweight='bold')
axes[2,0].set_title('Task 1: Wireless Delay (Dramatic Increase)', fontweight='bold', color='red')
axes[2,0].grid(True, alpha=0.3)
for i, (x, y) in enumerate(zip(wireless_hosts, wireless_delay)):
    axes[2,0].annotate(f'{y:.1f}ms', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

# Wired Delay
axes[2,1].plot(range(1, 7), wired_delay, 'b-s', linewidth=3, markersize=8)
axes[2,1].set_xlabel('Configuration Number', fontweight='bold')
axes[2,1].set_ylabel('End-to-End Delay (ms)', fontweight='bold')
axes[2,1].set_title('Task 3: Wired Delay (Minimal Increase)', fontweight='bold', color='blue')
axes[2,1].grid(True, alpha=0.3)
for i, y in enumerate(wired_delay):
    axes[2,1].annotate(f'{y:.3f}ms', (i+1, y), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()

# Summary comparison table
print("=== SIDE-BY-SIDE COMPARISON SUMMARY ===")
print()
print("WIRELESS (Task 1) vs WIRED (Task 3):")
print("-" * 60)
print(f"{'Metric':<20} {'Wireless':<20} {'Wired':<20}")
print("-" * 60)
print(f"{'Best PDR':<20} {'97.5%':<20} {'97.5%':<20}")
print(f"{'Worst PDR':<20} {'48.0%':<20} {'97.5%':<20}")
print(f"{'Max Throughput':<20} {'0.154 Mbps':<20} {'0.312 Mbps':<20}")
print(f"{'Min Delay':<20} {'8.9 ms':<20} {'0.098 ms':<20}")
print(f"{'Max Delay':<20} {'327.2 ms':<20} {'2.687 ms':<20}")
print(f"{'Saturation':<20} {'Yes (~200 hosts)':<20} {'No':<20}")
print("-" * 60)
```
